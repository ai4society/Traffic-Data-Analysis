```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming total2018 is your dataframe
# Group by 'County Category' and calculate the total cost
average_total_cost_per_category = total2018.groupby('County Category')['total'].sum().reset_index()

# Calculate the number of collisions per category
collisions_per_category = total2018['County Category'].value_counts().reset_index()
collisions_per_category.columns = ['County Category', 'Number of Collisions']

# Ensure both dataframes are sorted by 'County Category'
average_total_cost_per_category = average_total_cost_per_category.sort_values('County Category')
collisions_per_category = collisions_per_category.sort_values('County Category')

# Merge the two dataframes to ensure correct alignment
merged_data = average_total_cost_per_category.merge(collisions_per_category, on='County Category')

# Set the style
sns.set(style="whitegrid")

category_colors = {
    'Urban': 'skyblue',
    'Rural': 'orange',
    'Mostly Rural': 'green'
}

# Create the bar plot
plt.figure(figsize=(7, 4))
bar_plot = sns.barplot(
    x='County Category', 
    y='total', 
    data=merged_data, 
    palette=category_colors
)

# Set plot labels and title
bar_plot.set_xlabel('County Category', fontsize=10)
bar_plot.set_ylabel('Total Cost (Log Scale)', fontsize=10)
bar_plot.set_title('Total Cost and Number of Collisions by County Category - 2018', fontsize=10)
bar_plot.set_yscale('log')

# Create a secondary y-axis
ax2 = bar_plot.twinx()

# Plot the line plot for the number of collisions
line_plot = sns.lineplot(
    x='County Category', 
    y='Number of Collisions', 
    data=merged_data, 
    ax=ax2, 
    color='red', 
    marker='o'
)

# Add text annotations for the number of collisions
for i, row in merged_data.iterrows():
    ax2.text(i, row['Number of Collisions'] + 5000, int(row['Number of Collisions']), 
             color='blue', ha="center", va="center", backgroundcolor='white', fontsize=8)

# Set the secondary y-axis labels
ax2.set_ylabel('Number of Collisions', fontsize=12)
# ax2.set_yscale('log')

plt.savefig('total_cost_and_collisions_by_county_category.png', bbox_inches='tight')

# Show the plot
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the total2018 dataframe
# Group by 'County of Collision' and calculate the mean of 'total'
average_total_cost_per_county = total2018.groupby('County of Collision')['total'].mean().reset_index()
average_total_cost_per_county['County Category'] = average_total_cost_per_county['County of Collision'].map(county_category_mapping)

# Set the style
sns.set(style="whitegrid")

# Create a color palette for the categories
category_colors = {
    'Urban': 'skyblue',
    'Rural': 'orange',
    'Mostly Rural': 'green'
}

# Create the bar plot with colored bars based on categories
plt.figure(figsize=(12, 7))
bar_plot = sns.barplot(
    x='County of Collision', 
    y='total', 
    data=average_total_cost_per_county, 
    hue='County Category', 
    dodge=False,
    palette=category_colors
)

# Set plot labels and title
bar_plot.set_xlabel('County of Collision', fontsize=12)
bar_plot.set_ylabel('Average Total Cost', fontsize=12)
bar_plot.set_title('Average Total Cost per County of Collision - 2018', fontsize=15)

# Rotate x-ticks for better readability if necessary
plt.xticks(rotation=90)

# Add the legend
plt.legend(title='County Category')

plt.savefig('average_cost_by_county.jpg', bbox_inches='tight')

# Show the plot
plt.show()

```
